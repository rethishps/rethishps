# Chapter 3, Exercise 4

Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.

The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:

## Exercise 6.a
Read in the covid19_cases.csv file.

```python
# importing the required libraries
import pandas as pd
```

```python
# Reading the CSV File
df_covid_19 = pd.read_csv('./covid19_data.csv')
df_covid_19.head(5)
```

## Exercise 6.b

Create a date column using the data in the dateRep column and the pd.to_datetime() function.

```python
# Adding a new date field by converting the dateRep column to date time format
df_covid_19['date'] = pd.to_datetime(df_covid_19['dateRep'], format='%d/%m/%Y')
df_covid_19.head(10)
```

## Exercise 6.c

Set the date column as the index and sort the index.

```python
# setting the date as the index column and sorting 
df_covid_19  = df_covid_19.set_index('date', drop=False).sort_index()
df_covid_19
```

## Exercise 6.d

Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively. Hint: the replace() method can be run on the dataframe as a whole.

```python
df_covid_19
```

```python
## Replacing the United_States_of_America and United_Kingdom with USA and UK

df_covid_19['countriesAndTerritories']=df_covid_19['countriesAndTerritories'].replace(
    'United_States_of_America', 'USA')

df_covid_19['countriesAndTerritories']=df_covid_19['countriesAndTerritories'].replace(
    'United_Kingdom', 'UK')

df_covid_19[df_covid_19['countriesAndTerritories'].isin(['USA','UK'])]

```

## Exercise 6.e

Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.

```python
# Filtering the data for the specific countries
df_covid_19_filtered = df_covid_19[df_covid_19['countriesAndTerritories'].isin(['USA','UK',
    'Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 
    'Spain', 'Turkey'])]
df_covid_19_filtered
```

## Exercise 6.f

Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). Be sure to fill in NaN values with 0.



```python
# Dropping the duplicate column date as date is index and another column
df_covid_19_filtered_datedroped = df_covid_19_filtered.drop('date', axis=1)
# Creating the pivot table
df_covid_19_filtered_pivot=df_covid_19_filtered_datedroped.pivot_table(
    index='date',columns=['countriesAndTerritories'],
    values=['cases'],margins=False).fillna(0) 
df_covid_19_filtered_pivot    
```

```python
# Pivot basic method
df_covid19_pivot=df_covid_19_filtered.pivot(index='date',columns=['countriesAndTerritories'],values=['cases']).fillna(0) 
df_covid19_pivot
```
