# Exercise 1.

With the earthquakes.csv file, select all the earthquakes in Japan with a magnitude of 4.9 or greater using the mb magnitude type.

```python
# Reading the earthquakes dataset
df_earthquakes = pd.read_csv('earthquakes.csv')
df_earthquakes.head()
```

```python
df_earthquakes_jpn = df_earthquakes[(df_earthquakes['place'].str.contains('Japan')) & (df_earthquakes['mag'] >= 4.9) & (df_earthquakes['magType'] == 'mb')]
df_earthquakes_jpn
```

# Exercise 2.

Create bins for each full number of earthquake magnitude (for instance, the first bin is (0,1], the second is (1,2], and so on) with the ml magnitude type and count how many are in each bin.



```python
# Filter for 'ml' magnitude type
df_ml = df_earthquakes[df_earthquakes['magType'] == 'ml']

# Create bins for each full number of magnitude
bins = range(0, int(df_ml['mag'].max()) + 2)  # +2 to include the last edge
print(bins)
print(df_ml['mag'].max())

# Creating the labels for the bins
labels = [f"({i},{i+1}]" for i in range(0, int(df_ml['mag'].max()) + 1)]
print(labels)
# Assigning bins to a new column using pd.cut
df_ml['mag_bin'] = pd.cut(df_ml['mag'], bins=bins, labels=labels, right=True)
df_ml.head()
# Count earthquakes in each bin
bin_counts = df_ml['mag_bin'].value_counts().sort_index()
print(bin_counts)
```

# Exercise 4.

Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magnitude type along the columns.

```python
# Creating the crosstab using the pandas crosstab function
pd.crosstab(index=df_earthquakes['tsunami'], columns=df_earthquakes['magType'],  \
            values=df_earthquakes['mag'], aggfunc='max')
```

# Exercise 6.

Create a pivot table of the FAANG data that compares the stocks. Put the ticker in the rows and show the averages of the OHLC and volume traded data.

# Exercise 7.

Calculate the z-scores for each numeric column of Amazon’s data (ticker is AMZN) in Q4 2018 using apply().

```python
# Filtering the Amazon data for Q4 2018
df_amzn_q4_2018 = df_faang[(df_faang['ticker'] == 'AMZN') & 
                           (df_faang['date'] >= '2018-10-01') & 
                           (df_faang['date'] <= '2018-12-31')]
# Applying the z-score formula using apply() for the OHLC and volume columns
z_scores_amzn = df_amzn_q4_2018[['open', 'high', 'low', 'close', 'volume']]. \
    apply(lambda x: (x - x.mean()) / x.std())
z_scores_amzn
```

The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data through September 18, 2020. Complete the following tasks to practice the skills you've learned up to this point in the book:

## 1. Prepare the data

### a. Read in the data in the covid19_cases.csv file.

```python
df_covid19 = pd.read_csv('covid19_cases.csv')
df_covid19.head()
```

## b. Create a date column by parsing the dateRep column into a datetime.

```python
# Adding a new column dateRep_dt by parsing the dateRep column into a datetime
df_covid19['dateRep_dt'] = df_covid19['dateRep'].\
    apply(lambda x: pd.to_datetime(x, format='%d/%m/%Y')) 
df_covid19.head()
```

### c. Set the date column as the index.

```python
# Setting the dateRep_dt column as the index using set_index() function
df_covid19.set_index('dateRep_dt', inplace=True)
df_covid19.head()
```

### d. Use the replace() method to update all occurrences of United_States_of_America and United_Kingdom to USA and UK, respectively.

```python
df_covid19.replace({'United_States_of_America': 'USA', 'United_Kingdom': 'UK'}, inplace=True)
```

```python
# Checking if the replacements were successful for USA
df_covid19[df_covid19['countriesAndTerritories'].isin(['USA'])].head()
```

```python
# Checking if the replacements were successful for UK
df_covid19[df_covid19['countriesAndTerritories'].isin(['UK'])].head()
```

## 2. For the five countries with the most cases (cumulative), find the day with the largest number of cases.

```python
df_covid19.head()
```

```python
# Finding the five countries with the most cumulative cases
top_five_coutries = df_covid19.groupby('countriesAndTerritories')['cases'].sum().nlargest(5).index
top_five_coutries
```

```python
# Find the day with the largest number of cases for these top five countries (identified above)
df_covid19[df_covid19['countriesAndTerritories'].isin(top_five_coutries)]. \
    groupby('countriesAndTerritories').cases.idxmax()
```

## 3. Find the 7-day average change in COVID-19 cases for the last week in the data for the five countries with the most cases.

```python
# Filtering the dataframe for only the top five countries
df_cv19_top5_countries = df_covid19[df_covid19['countriesAndTerritories'].isin(top_five_coutries)]
df_cv19_top5_countries.countriesAndTerritories.unique()
```

```python
df_cv19_top5_countries_grp = df_cv19_top5_countries.groupby(['countriesAndTerritories', pd.Grouper(freq='1d')]).cases.sum()
df_cv19_top5_countries_grp
```

```python
# Unstacking the grouped dataframe, 
# calculating the daily difference, 
# applying a 7-day rolling mean, 
# and filtering for the last week

df_cv19_top5_countries_grp.unstack(0).diff().rolling(7).mean().last('1W')
```

## 4. Find the first date that each country other than China had cases.

```python
# Filtering out China and removing the rows with zero cases,
# Finding the first date that each country had cases
df_covid19[(df_covid19['countriesAndTerritories'] !='China') & (df_covid19["cases"]!=0) ].\
    groupby('countriesAndTerritories').\
    cases.idxmin()
```

## 5. Rank the countries by cumulative cases using percentiles.

```python
# Grouping by countries and ranking by cumulative cases using percentiles

df_covid19_rank=df_covid19.groupby("countriesAndTerritories").cases.sum().rank(pct=True)
df_covid19_rank
#sort the data in descending order
df_covid19_rank.sort_values(ascending=False)
```
