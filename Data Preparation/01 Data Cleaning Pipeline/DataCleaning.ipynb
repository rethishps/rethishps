```python
# importing necessary libraries
import pandas as pd
import numpy as np
```

## Chapter 7

```python
df_candy = pd.read_excel('./data/candyhierarchy2017.xlsx')
df_candy.head()
```

```python
df_candy.shape
```

```python
# Filter out missing data dropna with how='all' means that only rows where all elements are missing will be dropped
df_candy_clean = df_candy.dropna(how='all')
df_candy_clean
# No rows with all missing values are present in the dataframe hence no rows are dropped
```

```python
## Filter out values whre a column has missing values
df_candy_q2_gender_clean = df_candy.dropna(subset=['Q2: GENDER'])
df_candy_q2_gender_clean.shape
print("Original DataFrame shape:", df_candy.shape)
print("DataFrame shape after dropping rows with missing values in 'Q2: GENDER':", df_candy_q2_gender_clean.shape)
print("DataFrame shape after dropping rows with missing 'Q2: GENDER' :", (df_candy.shape[0] - df_candy_q2_gender_clean.shape[0]), "rows removed.")  
```

```python
# Remove duplicates
df_candy_no_duplicates = df_candy.drop_duplicates()
print("Original DataFrame shape:", df_candy.shape)
print("DataFrame shape after removing duplicates:", df_candy_no_duplicates.shape)       
# No Duplicates found in the dataset
```

```python
# Manipulate Strings - Convert Country to Upper Case
df_candy['Q4: COUNTRY'] = df_candy['Q4: COUNTRY'].str.upper()
df_candy.head()
```

```python
df_candy.head()
```

```python
# Convert between string and date time

df_candy_2016['Timestamp'] = pd.to_datetime(df_candy_2016['Timestamp'])
df_candy_2016.head()
```

```python
# Convert timestamps to periods and back
# To convert a Timestamp to a Period, use the to_period() method, specifying the desired frequency.
df_candy_2016['Period'] = df_candy_2016['Timestamp'].dt.to_period('M')
df_candy_2016.head()
```
