## Calculate :

Calculate some useful statistics, such as mean rooms per dwelling, median age, mean distances to five Boston employment centers, and the percentage of houses with a low price (< $20,000).

```python
# Mean rooms per dwelling
#print(df_boston_housing_new.columns.to_list())
print('Mean rooms per dwelling ',df_boston_housing_new['RM'].mean())

# Median Age
print('Median Age ',df_boston_housing_new['AGE'].median())

# Mean distance 
print('Mean Distance ',df_boston_housing_new['DIS'].mean())

# Calculate percentage of house with low price ( $20K )
low_price_houses = df_boston_housing_new[df_boston_housing_new.PRICE < 20]

# Calculating the percentage value 
perc = ( (len(low_price_houses.PRICE) / len(df_boston_housing_new.PRICE)) * 100)

print('Percentage of house with <20,000 price is:' , perc)

```

## 8. Create a function to strip the whitespace characters.

```python
def strip_me(str_val):
    return str_val.strip()
```

## 9. Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column, and drop the new column. 

```python
df_adult_income = pd.read_csv('./adult_income_data.csv',names=columns_head)
df_adult_income.shape
```

```python
str_cols = df_adult_income.select_dtypes(exclude=['int64']).columns
str_cols.to_list()

for col in str_cols.to_list():
    print(col)
    df_adult_income[col+'_stripped'] = df_adult_income[col].apply(strip_me)
    df_adult_income = df_adult_income.drop(col, axis=1)
    df_adult_income.rename(columns={col+'_stripped':col}, inplace=True)


df_adult_income
```
