## 1. What is the count of people belonging to different age groups in the persons table?

```python
conn = sqlite3.connect("petsdb")
c = conn.cursor()
```

```python
# Checking the columns of persons table

c.execute("SELECT * FROM persons limit 5")
column_names = [description[0] for description in c.description]

print(column_names)
for row in c.fetchall():
    print(row)
```

```python
# Using the cursor object to iterate through the query
# The SQL is using an aggreration function count over age of different people

for ppl, age in c.execute("SELECT count(*), age FROM persons GROUP BY age"):
    print("We have {} people aged {}".format(ppl, age))
```

## 2. Which age group has the maximum number of people?

```python
# Here we are using the sorting function using th order by clause
# The loop breaks after priting the first row
for ppl, age in c.execute(
    "SELECT count(*), age FROM persons GROUP BY age ORDER BY count(*) DESC"):
    print("The highest number of people is {} and came from {} age group".format(ppl, age))
    break
```

## 4. How many people have more than one pet?

```python
# This is step we are just printing the Peorples having more than one pet
c.execute(" select ps.id, ps.first_name,  count(*) pet_count from persons ps, pets pt where ps.id = pt.owner_id " \
            " group by  ps.id, ps.first_name having count(*) > 1 order by pet_count desc")

rows =c.fetchall()

for r in rows:
    print("Owner Id : {} Onwer Name : {}  Pet Count : {} ".format(r[0],r[1],r[2]))
```

```python
## This statement will print the count of People having more than one pet

c.execute(" select count(*) from ( " \
            "select ps.id, ps.first_name,  count(*) pet_count " \
            " from persons ps, pets pt where ps.id = pt.owner_id " \
            " group by  ps.id, ps.first_name having count(*) > 1 ) ")

rows =c.fetchall()

print("Total Number of Persons Having morethan one pet ", rows[0][0])
```
